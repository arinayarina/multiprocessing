```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
s2 = '0.466825 0.485340 1000000 220000 1 \
0.466825 0.293069 1000000 220000 2 \
0.466825 0.291055 1000000 220000 4'

X2 = s2.split()
T = np.zeros((0, 1))
T_qs = np.zeros((0, 1))
P = np.zeros((0, 1))
i = 0
while i < len(X2):
    if i % 5 == 0:
        T_qs = np.append(T_qs, float(X2[i]))
    if i % 5 == 1:
        T = np.append(T, float(X2[i]))
    if i % 5 == 4:
        P = np.append(P, float(X2[i]))
    i+=1
    
S = np.zeros((0, 1))
E = np.zeros((0, 1))
i = 0
while i < len(P):
    S = np.append(S, T[0]/T[i])
    E = np.append(E, S[i]/P[i])
    i+=1
```

```python
plt.plot(P, T, label='mergesort')
plt.plot(P, T_qs, label='qsort')
plt.title(u'Зависимость времени работы программы от числа потоков T(P)')
plt.legend()
plt.show()

plt.plot(P, S)
plt.title(u'Зависимость ускорения от числа потоков S(P)')
plt.show()

plt.plot(P, E)
plt.title(u'Зависимость эффективности от числа потоков E(P)')
plt.show()
```
