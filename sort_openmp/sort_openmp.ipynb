Вывод: можно выбрать размер чанка в размере 20% от общего числа элементов массива.

```python
s2 = '0.272665 0.460225 1000000 220000 1 \
0.240450 0.130420 1000000 220000 2 \
0.245570 0.182822 1000000 220000 4 \
0.238491 0.135342 1000000 220000 8 \
0.238565 0.143351 1000000 220000 16'
X2 = s2.split()
T = np.zeros((0, 1))
T_qs = np.zeros((0, 1))
P = np.zeros((0, 1))
i = 0
while i < len(X2):
    if i % 5 == 0:
        T_qs = np.append(T_qs, float(X2[i]))
    if i % 5 == 1:
        T = np.append(T, float(X2[i]))
    if i % 5 == 4:
        P = np.append(P, float(X2[i]))
    i+=1
    
S = np.zeros((0, 1))
E = np.zeros((0, 1))
i = 0
while i < len(P):
    S = np.append(S, T[0]/T[i])
    E = np.append(E, S[i]/P[i])
    i+=1
```

```python
plt.plot(P, T, label='mergesort')
plt.plot(P, T_qs, label='qsort')
plt.title(u'Зависимость времени работы программы от числа потоков T(P)')
plt.legend()
plt.show()

plt.plot(P, S)
plt.title(u'Зависимость ускорения от числа потоков S(P)')
plt.show()

plt.plot(P, E)
plt.title(u'Зависимость эффективности от числа потоков E(P)')
plt.show()
```
